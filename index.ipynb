**4. Compute the Z-score for a tree yielding 35 pounds of apples. Interpret this Z-score.**

```python
def calculate_z(x, mu, sigma):
    z = (x - mu) / sigma
    return z

calculate_z(35, 42.4, 6.0)
```

```python
# Interpret the result

# a tree with a yield of 35 pounds has a z-score of -1.233.
# Meaning that it's yield differs from the mean by -1.233 standard deviations.
```

**5. Suppose a tree has a Z-score of 1.85. Interpret this Z-score. What is the yield of this tree?**


```python
# Interpret the z score
```

```python
# Calculate yield
def return_yield(z, mu, sigma):
    x = z*sigma + mu
    return x

return_yield(1.85, 42.4, 6)
```

```python
# What is the yield ?
# 53.5 pounds
```
